```python
import numpy as np
import pandas as pd
```

```python
# Load dataset
dataset = pd.read_csv(r"C:\Users\Md Shah Alak\Downloads\datset - Sheet1.csv")
```

```python
dataset.head()
```

```python
# Clean missing/null values
cleaned_df = dataset.dropna()
```

```python
cleaned_df
```

```python
# Sum of sales by region
sales_by_region = cleaned_df.groupby('Region')['TotalSales'].sum()
print("Sum of Sales by Region:\n", sales_by_region, "\n")
```

```python
# Average sales per product
avg_sales_per_product = cleaned_df.groupby('Product')['TotalSales'].mean()
print("Average Sales per Product:\n", avg_sales_per_product, "\n")
```

```python
# Highest & lowest selling products
total_sales_by_product = cleaned_df.groupby('Product')['TotalSales'].sum()
highest_selling_product = total_sales_by_product.idxmax()
lowest_selling_product = total_sales_by_product.idxmin()
print("Highest Selling Product:", highest_selling_product)
print("Lowest Selling Product:", lowest_selling_product, "\n")
```

```python
# NumPy statistics for numerical fields
numerical_fields = cleaned_df.select_dtypes(include=[np.number])

mean_values = numerical_fields.mean()
median_values = numerical_fields.median()
std_deviation = numerical_fields.std()
```

```python
print("Mean of Numerical Fields:\n", mean_values, "\n")
print("Median of Numerical Fields:\n", median_values, "\n")
print("Standard Deviation of Numerical Fields:\n", std_deviation)
```
